```python
filtered_cdmx = incidencia_cdmx[incidencia_cdmx.longitud <= -20]
filtered_cdmx = filtered_cdmx[filtered_cdmx.calle_hech != '.']
```

```python
plt.figure(figsize=(7.5,10))
plt.scatter(
    x=filtered_cdmx.longitud,
    y=filtered_cdmx.latitud
)
```

<img src="Files/cdmx_map.jpg" />

```python
count_cdmx = filtered_cdmx.groupby(by=['coloniahec', 'alcaldiahe']).count()
first_cdmx = filtered_cdmx.sort_values(by=['longitud', 'latitud']).groupby(by=['coloniahec', 'alcaldiahe']).first()
```

```python
coord_cdmx = pd.concat(
    [
        count_cdmx['gid'],
        first_cdmx['latitud'],
        first_cdmx['longitud']
    ],
    axis=1
).reset_index().sort_values(by=['gid'], ascending=True)
```

```python
coord_cdmx.tail(1)
```

```python
plt.figure(figsize=(7.5,10))
plt.rcParams['image.cmap'] = 'coolwarm'
plt.scatter(
    x=coord_cdmx.longitud,
    y=coord_cdmx.latitud,
    c=np.log(coord_cdmx.gid),
    s=np.power(coord_cdmx.gid, 1/np.sqrt(2))
)
```

```python
count_cdmx_1 = filtered_cdmx.groupby(by=['calle_hech', 'coloniahec', 'alcaldiahe']).count()
first_cdmx_1 = filtered_cdmx.sort_values(
    by=['longitud', 'latitud']
).groupby(by=['calle_hech', 'coloniahec', 'alcaldiahe']).first()
```

```python
coord_cdmx_1 = pd.concat(
    [
        count_cdmx_1['gid'],
        first_cdmx_1['latitud'],
        first_cdmx_1['longitud']
    ],
    axis=1
).reset_index().sort_values(by=['gid'], ascending=True)
```

```python
plt.figure(figsize=(7.5,10))
plt.scatter(
    x=coord_cdmx.longitud,
    y=coord_cdmx.latitud,
    c=np.log(coord_cdmx.gid),
    s=np.power(coord_cdmx.gid, 1/np.sqrt(2))
)
plt.scatter(
    x=coord_cdmx_1.longitud,
    y=coord_cdmx_1.latitud,
    c=np.log(coord_cdmx_1.gid),
    s=np.power(coord_cdmx_1.gid, 1/np.sqrt(2))
)
```

<h1>Análisis temporal</h1>

```python
count_cdmx_2 = filtered_cdmx.groupby(
    by=['ano_hecho', 'mes_hecho']
).count().reset_index()
```

```python
count_cdmx_2.mes_hecho.replace('Enero', '01', inplace=True)
count_cdmx_2.mes_hecho.replace('Febrero', '02', inplace=True)
count_cdmx_2.mes_hecho.replace('Marzo', '03', inplace=True)
count_cdmx_2.mes_hecho.replace('Abril', '04', inplace=True)
count_cdmx_2.mes_hecho.replace('Mayo', '05', inplace=True)
count_cdmx_2.mes_hecho.replace('Junio', '06', inplace=True)
count_cdmx_2.mes_hecho.replace('Julio', '07', inplace=True)
count_cdmx_2.mes_hecho.replace('Agosto', '08', inplace=True)
count_cdmx_2.mes_hecho.replace('Septiembre', '09', inplace=True)
count_cdmx_2.mes_hecho.replace('Octubre', '10', inplace=True)
count_cdmx_2.mes_hecho.replace('Noviembre', '11', inplace=True)
count_cdmx_2.mes_hecho.replace('Diciembre', '12', inplace=True)

count_cdmx_2 = count_cdmx_2[count_cdmx_2.ano_hecho >= 2018].sort_values(by=['ano_hecho', 'mes_hecho'])
```

```python
time_cdmx = pd.concat(
    [
        count_cdmx_2['gid'],
        count_cdmx_2['ano_hecho'],
        count_cdmx_2['mes_hecho']
    ],
    axis=1
).reset_index()
```
