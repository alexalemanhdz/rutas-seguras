```python
esquinas_2 = esquinas.copy()
```

```python
esquinas_2
```

```python
esquinas_2['Calle 1'] = esquinas_2['Calle 1'].map(lambda x: str(x).upper())
esquinas_2['Calle 2'] = esquinas_2['Calle 2'].map(lambda x: str(x).upper())
esquinas_2['Calle 1'] = esquinas_2['Calle 1'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
esquinas_2['Calle 2'] = esquinas_2['Calle 2'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```python
esquinas_2
```

```python
esquinas_2['Calle 1'] = esquinas_2['Calle 1'].replace('CALLE ZOLAPAN', 'ZOLAPAN')
esquinas_2['Calle 2'] = esquinas_2['Calle 2'].replace('CALLE ZOLAPAN', 'ZOLAPAN')
esquinas_2['Calle 1'] = esquinas_2['Calle 1'].replace('CDA. CHALCHICOMULA', 'CERRADA DE CHALCHICOMULA')
esquinas_2['Calle 2'] = esquinas_2['Calle 2'].replace('CDA. CHALCHICOMULA', 'CERRADA DE CHALCHICOMULA')

unique_street_names = pd.unique(esquinas_2[['Calle 1', 'Calle 2']].values.ravel('K'))
```

```python
unique_street_names
```

```python
incidences_clean = incidencias_pedregal[incidencias_pedregal.calle_hech != '0'].calle_hech

incidences_clean = incidences_clean.replace('CHACHICOMULA', 'CHALCHICOMULA')
incidences_clean = incidences_clean.replace('Y TEJAMANIL', 'TEJAMANIL')
incidences_clean = incidences_clean.replace('COACOYUAGA', 'COACOYUNGA')
incidences_clean = incidences_clean.replace('XOHIAPAN', 'XOCHIAPAN')
incidences_clean = incidences_clean.replace('DELFIN MADRIGAL', 'AV. ANTONIO DELFIN MADRIGAL')
incidences_clean = incidences_clean.replace('ANACAHUITA', 'AV. ANACAHUITA')
incidences_clean = incidences_clean.replace('AV DE LAS ROSAS', 'TEPEYAUCLE')
incidences_clean = incidences_clean.replace('ESQUINA PAPALOTL', 'CANAHUTLI')
incidences_clean = incidences_clean.replace('ESCUINAPA', 'AV. ESCUINAPA')
incidences_clean = incidences_clean.replace('PEDRO HENRIQUEZ UREÑA', 'AV. PEDRO HENRIQUEZ URENA')
```

```python
incidence_streets = pd.DataFrame(
    data= {
        'calle': incidences_clean,
        'in_esquinas': incidences_clean.isin(unique_street_names),
        'latitud': incidencias_pedregal.latitud,
        'longitud': incidencias_pedregal.longitud
    }
)
```

```python
street_series = pd.DataFrame(unique_street_names, columns=['street'])
```

```python
street_series
```

```python
calles_1 = street_series.street.map(lambda x: esquinas_2[esquinas_2['Calle 1'] == x]['Calle 2'])
calles_2 = street_series.street.map(lambda x: esquinas_2[esquinas_2['Calle 2'] == x]['Calle 1'])
```

```python
calles_3 = pd.DataFrame()

for index, value in calles_1.items():
    calles_3 = pd.concat([calles_3, pd.concat([value, calles_2[index]], axis=0).reset_index(drop=True)], axis=1)
```

```python
crossings = calles_3.transpose().reset_index(drop=True).transpose()
```

```python
crossings
```

```python
incidencies_count = pd.DataFrame(incidencias_pedregal.groupby('calle_hech').count().gid).reset_index()
```

```python
incidencies_count
```

```python
street_series['crime_index'] = street_series.street.map(
    lambda x: incidencies_count[incidencies_count['calle_hech'] == x].values[0][-1] if incidencies_count[incidencies_count['calle_hech'] == x].size > 0 else 0
)
```

```python
street_series.head(10)
```

```python
start_street = 'CHALCHICOMULA'
end_street = 'DALIAS'
current = start_street
unvisited = {node: None for node in unique_street_names}
visited = {}
current_distance = {'distance': 0, 'predecessor': None}

while True:
    index = street_series.index[street_series.street == current][0]
    for value in crossings[index].values:
        if isinstance(value, str):
            if value not in unvisited: continue
            crime_index = street_series[street_series.street == value].values[0][1]
            new_distance = current_distance["distance"] + crime_index
            if unvisited[value] is None or unvisited[value]["distance"] > new_distance:
                unvisited[value] = {'distance': new_distance, 'predecessor': current}
    visited[current] = current_distance
    del unvisited[current]
    if not unvisited: break
    candidates = [node for node in unvisited.items() if node[1] is not None]
    current, current_distance = sorted(candidates, key = lambda x: x[1]["distance"])[0]
```

```python
current_back = end_street
while current_back != start_street:
    print(current_back, visited[current_back])
    current_back = visited[current_back]["predecessor"]
print(start_street, visited[start_street])
```
