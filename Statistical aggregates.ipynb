```python
filtered_cdmx = incidencia_cdmx[incidencia_cdmx.longitud <= -20]
```

```python
count_cdmx = filtered_cdmx.groupby(by=['coloniahec']).count()
first_cdmx = filtered_cdmx.sort_values(by=['longitud', 'latitud']).groupby(by=['coloniahec']).first()
```

```python
coord_cdmx = pd.concat([count_cdmx['gid'], first_cdmx['latitud'], first_cdmx['longitud']], axis=1).reset_index()
```

```python
coord_cdmx = coord_cdmx.sort_values(by=['gid'], ascending=True)
```

```python
plt.rcParams['image.cmap'] = 'coolwarm'
```

```python
plt.scatter(x=coord_cdmx.longitud, y=coord_cdmx.latitud, c=np.log(coord_cdmx.gid), s=np.sqrt(coord_cdmx.gid))
```
